```python
#null values
null_values = data.isnull().sum()
null_values
```

```python
#nilai unik
unique_values = data.nunique()
unique_values
```

```python
#Visualisai nilai unik Bar Chart
sns.barplot(x=unique_values.index, y=unique_values.values, palette="pastel")
plt.xticks(rotation=90)
plt.xlabel('Kolom')
plt.ylabel('Jumlah Nilai Unik')
plt.title('Jumlah Nilai Unik per Kolom')
plt.show()
```

```python
#filter berdasarkan key 'A'
filtered1_data = data[data['key'] == 'A']
filtered1_data
```

```python
df_graphics = data[['artist_count', 'released_year', 'released_month', 'released_day', 
                    'in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists', 
                    'in_apple_charts', 'in_deezer_charts', 'bpm', 'danceability_%', 
                    'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 
                    'liveness_%', 'speechiness_%']]
plt.figure(figsize=(15, 8))

# Membuat boxplot untuk setiap kolom numerik
for i, col in enumerate(df_graphics.select_dtypes(include=['float64', 'int64']).columns):
    plt.subplot(1, len(df_graphics.select_dtypes(include=['float64', 'int64']).columns), i + 1)
    plt.boxplot(df_graphics[col])  # Mengabaikan nilai NaN
    plt.title(col)
    plt.xlabel("Nilai")
    
plt.suptitle("Boxplot untuk Semua Kolom Numerik", fontsize=16)
plt.tight_layout()
plt.show()

#ambil yg integer
```

```python
# Memilih hanya kolom numerik untuk perhitungan korelasi
numerical_df = data.select_dtypes(include=[np.number])

# Menampilkan korelasi antar variabel numerik (Heatmap)
plt.figure(figsize=(10,6))
sns.heatmap(numerical_df.corr(), annot=True,cmap='coolwarm' , linewidths=0.5)
plt.title('Heatmap Korelasi Antar Variabel Numerik')
plt.show()
```

```python
data.hist(bins=20, figsize=(10,8), grid=False)
plt.tight_layout()
plt.show()
```

```python
# Memilih kolom numerik
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Menghitung matriks korelasi
correlation_matrix = numeric_data.corr()

# Menampilkan matriks korelasi
correlation_matrix

# Membuat heatmap untuk visualisasi korelasi
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Menambahkan judul
plt.title('Correlation Heatmap')

# Menampilkan plot
plt.show()
```
